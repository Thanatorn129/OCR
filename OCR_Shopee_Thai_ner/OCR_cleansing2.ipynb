```python
def censor_name(df, column_name, censor_char='*'):
    censored_df = df.copy()
    censored_df[column_name] = censored_df[column_name].apply(
        lambda x: ' '.join(
            [word if index != 2 else censor_char * len(word) for index, word in enumerate(x.split())]
        )
    )
    return censored_df
```

```python
def fix_span_error(words,ner):
    _ner = []
    _ner=ner
    _new_tag=[]
    for i,j in zip(words,_ner):
        #print(i,j)
        i=tokenizer.decode(i)
        if i.isspace() and j.startswith("B-"):
            j="O"
        if i=='' or i=='<s>' or i=='</s>':
            continue
        if i=="<_>":
            i=" "
        _new_tag.append((i,j))
    return _new_tag
```

```python
def merge_person_names(tags):
    merged_names = []
    current_name = ""
    inside_name = False

    for tag in tags:
        word, entity = tag
        if entity == "B-PERSON" or entity == "I-PERSON":
            if current_name != "":
                merged_names.append(current_name)
            current_name = word
            inside_name = True
        else:
            if inside_name:
                merged_names.append(current_name)
                current_name = ""
                inside_name = False

    if current_name != "":
        merged_names.append(current_name)

    return "".join(merged_names)
```

```python
from transformers import AutoTokenizer
from transformers import AutoModelForTokenClassification
from pythainlp.tokenize import word_tokenize # pip install pythainlp
import torch
for index, row in train.iterrows():
    name="pythainlp/thainer-corpus-v2-base-model"
    tokenizer = AutoTokenizer.from_pretrained(name)
    model = AutoModelForTokenClassification.from_pretrained(name)

    sentence = row.to_string(index=False)
    cut=word_tokenize(sentence.replace(" ", "<_>"))
    inputs=tokenizer(cut,is_split_into_words=True,return_tensors="pt")

    ids = inputs["input_ids"]
    mask = inputs["attention_mask"]
    # forward pass
    outputs = model(ids, attention_mask=mask)
    logits = outputs[0]

    predictions = torch.argmax(logits, dim=2)
    predicted_token_class = [model.config.id2label[t.item()] for t in predictions[0]]

    data=fix_span_error(inputs['input_ids'][0],predicted_token_class)
    person_names = merge_person_names(data)
    arrvalues.append(person_names)
```

```python
def censor_last_name(data, column_to_censor, censor_char='*****'):
    censored_data = data.copy()
    censored_data[column_to_censor] = censored_data[column_to_censor].apply(lambda x: x.split()[0] + ' ' + censor_char * (len(x.split()) - 1))
    return censored_data #censor real Name
```
